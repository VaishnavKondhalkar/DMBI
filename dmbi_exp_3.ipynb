```python
import pandas as pd

df = pd.read_csv('airtel.csv')

print(df.to_string()) 
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df = df.drop(['Close Price'], axis=1,  inplace=True)
```

```python
df_new=pd.get_dummies(df, columns=["Date"], prefix="Repeat_Dates")
```

```python
print(df_new)
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()

df_norm = df[['High Price','Low Price','No. of Trades','WAP']]
df_minmax = min_max_scaler.fit_transform(df_norm)
```

```python
pd.DataFrame(df_minmax)
```

```python
shares_status = df['WAP'].value_counts(ascending=False)
```

```python
shares_status
```

```python
len(df.WAP.unique())
```

```python

WAP_stats_less_than_500 = shares_status[shares_status<=500]
```

```python
WAP_stats_less_than_500
```

```python
df.WAP = df.WAP.apply(lambda x: 'other' if x in WAP_stats_less_than_500 else x)
```
